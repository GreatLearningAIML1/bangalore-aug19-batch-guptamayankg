## About the dataset

This is perhaps the best known database to be found in the pattern recognition literature. The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant.  One class is linearly separable from the other 2; the latter are NOT linearly separable from each other.

Attribute Information:
   1. sepal length in cm
   2. sepal width in cm
   3. petal length in cm
   4. petal width in cm
   5. class: 
      -- Iris Setosa

      -- Iris Virginica

### Plot the attrition column (Since it is a categorical variable, chose the plot accordingly)

```python
#Importing matplotlib.pyplot library 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
#Plotting Seaborn chart for Categorical Data i.e. Attrition
plt.figure(figsize=(8,6))
sx = sns.countplot(x='Attrition',data=df_hrmaster)
sx.set_title('Count of Attrition (Yes/No)')
sx.set_ylabel('Number of Employees')
plt.show()
```

### Plot Total working years using matplotlib plot function

```python
#Plotting chart for Years of experience of employees using matplotlib
df_hrmaster["TotalWorkingYears"].plot.hist()
plt.xlabel("Years of Experience") 
plt.ylabel("Number of Employees")
plt.title("Total Years of Work Experience")
plt.show()
```

```python
#Plotting chart for Years of experience of employees distribution using SeaBorn
s = sns.distplot(df_hrmaster["TotalWorkingYears"], kde=False)
s.set_title("Total Years of Work Experience Distribution")
s.set_ylabel("Number of Employees")
s.set_xlabel("Work Experience")
plt.show()
```
